```python
import pandas as pd  
import csv
import re
```

```python
df = pd.read_csv("ikeamaster_ver7.csv",sep=',',encoding='utf-8')

colors = df["summary"]
colors
```

```python
def column_filter(s):
    return s.split(',')

color_list = []

for n in range(len(colors)):
    
    data = column_filter(colors[n])
    
    if len(data) >= 2:
        
        color_list.append(data[1].strip())
    
    else:
        
        color_list.append(data[0].strip())    

print(color_list)
```

```python
co_1 = []
co_2 = []
co_3 = []
co_4 = []

for c in range(len(color_list)):
    color_data = re.sub('[a-zA-Z]', '', color_list[c]).split('/')    
    print('index:', c)
    print(color_data)   
    print('len:', len(color_data))
    print('='*20)

    if len(color_data) == 1:
        co_1.append(color_data[0])
        co_2.append('None')
        co_3.append('None')
        co_4.append('None')

    elif len(color_data) == 2:
        co_1.append(color_data[0])
        co_2.append(color_data[1])
        co_3.append('None')
        co_4.append('None')

    elif len(color_data) == 3:
        co_1.append(color_data[0])
        co_2.append(color_data[1])
        co_3.append(color_data[2])
        co_4.append('None')

    elif len(color_data) == 4:
        co_1.append(color_data[0])
        co_2.append(color_data[1])
        co_3.append(color_data[2])
        co_4.append(color_data[3])

print('co_1_length:' , len(co_1))
print('co_1:' ,co_1)
print('\n')
print('co_2_length:' , len(co_2))
print('co_2:' ,co_2)
print('\n')
print('co_3_length:' , len(co_3))
print('co_3:' ,co_3)
print('\n')
print('co_4_length:' , len(co_4))
print('co_4:' ,co_4)
print('\n')
  
```

```python
# color_1 去除多餘顏色資料，只留第二個

col_1 = []

for a in range(len(co_1)):
    
    if ' ' in co_1[a]:
        c1 = co_1[a].split(' ')[1]
        col_1.append(c1)
            
    else:
        col_1.append(co_1[a])

print(len(col_1))
print(col_1)
        
```

```python
# color_1 若為'', 補入color_2顏色

for b in range(len(col_1)):
    
    if col_1[b] == '':
        col_1[b].replace('', co_2[b])
#         col_1[b] = col_1[b].replace('', co_2[b])
               
    else:
        continue

print(len(col_1))
print(col_1)
```

```python
# color_2 去除多餘顏色資料，只留第二個

col_2 = []

for d in range(len(co_2)):
    
    if ' ' in co_2[d]:
        c2 = co_2[d].split(' ')[1]
        col_2.append(c2)
            
    else:
        col_2.append(co_2[d])

print(len(col_2))
print(col_2)
```

```python
# color_2 若有為'', 補入color_ˇ顏色

for e in range(len(col_2)):
    
    if col_2[e] == '':
        col_2[e].replace('', co_3[e])
              
    else:
        continue

print(len(col_2))
print(col_2)
```
